```python
#ngram

from nltk.tokenize import word_tokenize
from nltk.util import ngrams
sen='This is a very good book to study'
ng=ngrams(sequence=nltk.word_tokenize(sen),n=2)
for i in ng:
    print(i)
```

```python
#Stemming
#Porterstemmer

from nltk.stem import PorterStemmer
data=PorterStemmer()
lst=['programs','reached','walking','running','jumped','eating','walking']
for i in lst:
    print(i ,":" ,data.stem(i))
```

```python
from nltk.stem import SnowballStemmer
from nltk.tokenize import word_tokenize
st=SnowballStemmer('english')
for i in lst:
    print(i,":",st.stem(i))
```

```python
#lemmatizing
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
lem=WordNetLemmatizer()
print("Rock :",lem.lemmatize("rocks"))
print("Reaches:",lem.lemmatize("reaches"))
print("walking",lem.lemmatize("walking"))
print("jumping:",lem.lemmatize('jumping'))
```

```python
#How to remove special characters
#[a-z]
#[A-Z]
#[0-9]
#Regular Expression is used for removing special characters
# "^" is used to remove all other charachers otherthan in the list
import re
str1='Luminar technolab @# is IT fi%nishng school 12345 ! located@ at kakkanad!'

str2=re.sub('[^a-zA-Z0-9]',' ',str1)
str2
```
